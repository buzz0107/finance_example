```python
%matplotlib inline

import numpy as np
import pandas as pd
import scipy.stats as ss
import matplotlib.pyplot as plt

#Black and Scholes
def d1(S, K, r, sigma, T,t):
    return (np.log(S/K) + (r + sigma**2 / 2) * (T-t))/(sigma * np.sqrt(T-t))

def d2(S, K, r, sigma, T,t):
    return (np.log(S / K) + (r - sigma**2 / 2) * (T-t))/ (sigma * np.sqrt(T-t))

def BlackScholes(type,S, K, r, sigma, T,t):
    if type=="C":
        return S * ss.norm.cdf(d1(S, K, r, sigma, T,t)) - K * np.exp(-r * (T-t)) * ss.norm.cdf(d2(S, K, r, sigma, T,t))
    else:
        return K * np.exp(-r * (T-t)) * ss.norm.cdf(-d2(S, K, r, sigma, T,t)) - S * ss.norm.cdf(-d1(S, K, r, sigma, T,t))

def delta(type, S, K, r, sigma, T,t):
    if type=="C":
        return ss.norm.cdf(d1(S, K, r, sigma, T,t))

def gamma(type, S, K, r, sigma, T,t):
    if type=="C":
        return ss.norm.pdf(d1(S, K, r, sigma, T,t))/(S*sigma*np.sqrt(T-t))

def vega(type, S, K, r, sigma, T,t):
    if type=="C":
        return S*ss.norm.pdf(d1(S, K, r, sigma, T,t))*np.sqrt(T-t)

def theta(type, S, K, r, sigma, T,t):
    if type=="C":
        return -(S*ss.norm.pdf(d1(S, K, r, sigma, T,t))*sigma)/(2*np.sqrt(T-t))-r*K*np.exp(-r*(T-t))*ss.norm.cdf(d2(S, K, r, sigma, T,t))
```

```python
S = 100.0
K = 100.0
r=0.01
sigma = 0.3
T = 1
t=0
Otype='C'


print "S0\tstock price at time 0:", S
print "K\tstrike price:", K
print "r\tcontinuously compounded risk-free rate:", r
print "sigma\tvolatility of the stock price per year:", sigma
print "T\ttime to maturity in trading years:", T


c_BS = BlackScholes(Otype,S, K, r, sigma, T,t)
print "c_BS\tBlack-Scholes price:", c_BS
```

```python
greeks_time=pd.DataFrame()

delta_time=[delta(Otype, S, K, r, sigma, T,x/100.) for x in xrange(t*100,T*100)]
greeks_time['delta_time']=delta_time

gamma_time=[gamma(Otype, S, K, r, sigma, T,x/100.) for x in xrange(t*100,T*100)]
greeks_time['gamma_time']=gamma_time

vega_time=[vega(Otype, S, K, r, sigma, T,x/100.) for x in xrange(t*100,T*100)]
greeks_time['vega_time']=vega_time

theta_time=[theta(Otype, S, K, r, sigma, T,x/100.) for x in xrange(t*100,T*100)]
greeks_time['theta_time']=theta_time

greeks_underlying=pd.DataFrame()

delta_underlying=[delta(Otype, x, K, r, sigma, T,t) for x in range(0,200)]
greeks_underlying['delta_underlying']=delta_underlying

gamma_underlying=[gamma(Otype, x, K, r, sigma, T,t) for x in range(0,200)]
greeks_underlying['gamma_underlying']=gamma_underlying

vega_underlying=[vega(Otype, x, K, r, sigma, T,t) for x in range(0,200)]
greeks_underlying['vega_underlying']=vega_underlying

theta_underlying=[theta(Otype, x, K, r, sigma, T,t) for x in range(0,200)]
greeks_underlying['theta_underlying']=theta_underlying

greeks_time.plot(y='delta_time')
greeks_time.plot(y='gamma_time')
greeks_time.plot(y='vega_time')
greeks_time.plot(y='theta_time')


greeks_underlying.plot(y='delta_underlying')
greeks_underlying.plot(y='gamma_underlying')
greeks_underlying.plot(y='vega_underlying')
greeks_underlying.plot(y='theta_underlying')
```
